```python
import pandas as pd

df = pd.read_csv('adult_data.csv', header=None,
                     names=['age', 'workclass', 'fnlwgt', 'education', 'education-num',
                            'marital-status', 'occupation', 'relationship', 'race', 'sex',
                            'capital-gain', 'capital-loss', 'hours-per-week', 'native-country',
                            'salary'])

df.head()
```

```python
# 1. How many people of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = df['race'].value_counts()

print("Number of each race:\n", race_count)

```

```python
# 2. What is the average age of men?
average_age_men = df[df['sex'] == 'Male']['age'].mean().round(1)
print("Average age of men:", average_age_men)

# 3. What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = (df['education'] == 'Bachelors').mean() * 100

# 4. What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
advanced_education = df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])
high_income_advanced_education = df[advanced_education]['salary'] == '>50K'
percentage_high_income_advanced = (high_income_advanced_education.mean() * 100).round(1)

# 5. What percentage of people without advanced education make more than 50K?
non_advanced_education = ~advanced_education
high_income_non_advanced_education = df[non_advanced_education]['salary'] == '>50K'
percentage_high_income_non_advanced = (high_income_non_advanced_education.mean() * 100).round(1)

# 6. What is the minimum number of hours a person works per week?
min_work_hours = df['hours-per-week'].min()

# 7. What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
min_hours_workers = df[df['hours-per-week'] == min_work_hours]
percentage_high_income_min_hours = ((min_hours_workers['salary'] == '>50K').mean() * 100).round(1)

# 8. What country has the highest percentage of people that earn >50K and what is that percentage?
country_earning_counts = df[df['salary'] == '>50K']['native-country'].value_counts()
country_counts = df['native-country'].value_counts()
highest_earning_country = (country_earning_counts / country_counts * 100).idxmax()
highest_earning_country_percentage = (country_earning_counts / country_counts * 100).max().round(1)

# 9. Identify the most popular occupation for those who earn >50K in India.
top_IN_occupation = df[(df['native-country'] == 'India') & (df['salary'] == '>50K')]['occupation'].mode()

print(f"Percentage with Bachelors degrees: {percentage_bachelors}%")
print(f"Percentage with higher education that earn >50K: {percentage_high_income_advanced}%")
print(f"Percentage without higher education that earn >50K: {percentage_high_income_non_advanced}%")
print(f"Min work time: {min_work_hours} hours/week")
print(f"Percentage of rich among those who work fewest hours: {percentage_high_income_min_hours}%")
print("Country with highest percentage of rich:", highest_earning_country)
print(f"Highest percentage of rich people in country: {highest_earning_country_percentage}%")
print("Top occupations in India:", top_IN_occupation)
```
